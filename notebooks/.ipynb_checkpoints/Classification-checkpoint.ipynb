```python
import re
import string
import nltk
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from keras.preprocessing.sequence import pad_sequences
from keras.layers import Embedding


df = pd.read_csv('../data/test_gender_fusion.csv')
# Defina a proporção de dados para teste (por exemplo, 0.2 para 20%)
test_size = 0.2

# Se você quiser resultados consistentes a cada execução, defina uma semente aleatória
random_state = 42

train_data, test_data = train_test_split(df, test_size=test_size, random_state=random_state, shuffle="True")



train_x = train_data["Description Tokenized"]
train_y = train_data.iloc[:, 7:18]
val_x = test_data["Description Tokenized"]
val_y = test_data.iloc[:, 7:18]
```

```python

train_y = train_y.astype("float32")
val_y = val_y.astype("float32")

max_len = 50
train_x = pad_sequences(train_x, maxlen=max_len, padding='post')
val_x = pad_sequences(val_x, maxlen=max_len, padding='post')
test_x = pad_sequences(test_x, maxlen=max_len, padding='post')


model = Sequential()

embedding_dim = 150
vocab_size = 10000  # Ajuste esse valor de acordo com o tamanho do seu vocabulário


# Adicione o restante das camadas do modelo (como no seu código original)

# Define the number of LSTM layers and the number of neurons in each layer
num_lstm_layers = 3
num_lstm_neurons = 128

# Define the activation function for the LSTM layers
activation_func = 'relu'

# Define the dropout rate to prevent overfitting
dropout_rate = 0.2

# Define the number of classes for the multiclassification problem
num_classes = 11

# Define the model architecture
model = Sequential()

max_len=50
embedding_dim=150
# Adicione a camada de Embedding
model.add(Embedding(input_dim=vocab_size, output_dim=embedding_dim, input_length=max_len))

# Add the first LSTM layer
model.add(LSTM(units=num_lstm_neurons, activation=activation_func, return_sequences=True, input_shape=(max_len, embedding_dim)))
model.add(Dropout(rate=dropout_rate))

# Add additional LSTM layers
for i in range(num_lstm_layers-2):
    model.add(LSTM(units=num_lstm_neurons, activation=activation_func, return_sequences=True))
    model.add(Dropout(rate=dropout_rate))

# Add the last LSTM layer
model.add(LSTM(units=num_lstm_neurons, activation=activation_func))
model.add(Dropout(rate=dropout_rate))

# Add a dense layer with softmax activation function for multiclassification
model.add(Dense(units=num_classes, activation='softmax'))

optimizer = 'adam'
loss_func = 'categorical_crossentropy'
metrics = ['accuracy']

# Compile the model
model.compile(optimizer=optimizer, loss=loss_func, metrics=metrics)


# Define the number of epochs and batch size for training
num_epochs = 10
batch_size = 32

# Train the model
history = model.fit(train_x, train_y, epochs=num_epochs, batch_size=batch_size, validation_data=(val_x, val_y), verbose=1)

# Evaluate the model on the test set
test_loss, test_acc = model.evaluate(test_x, test_y, verbose=1)

# Print the test accuracy
print('Test accuracy:', test_acc)
```
