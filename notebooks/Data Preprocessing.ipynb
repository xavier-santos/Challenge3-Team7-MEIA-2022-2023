```python
import pandas as pd
import numpy as np
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import RegexpTokenizer
import string
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('stopwords')
```

```python
df = pd.read_csv('../data/movies_database.csv', na_values=['#EMPTY?'], delimiter=";")
```

```python
df = df.dropna(subset="Description")
```

```python
df.isnull().sum()
```

```python
df["Description"].unique().size
```

```python
df = df.drop_duplicates(subset="Description")
```

```python
df["Name"].size
```

```python
dummies = df["Genre(s)"].str.get_dummies(sep=",")
```

```python
df = pd.concat([df, dummies], axis=1)
```

```python
pd.set_option("display.max_columns", None)
df.head(5)
```

```python
df['Keywords'] = df['Keywords'].fillna(" ")
df['Keywords'] = df["Keywords"].str.split(",")
df.head(5)
```

```python
tokenizer = RegexpTokenizer(r'[a-zA-z]+')
df['Description Tokenized'] = df.apply(lambda row: tokenizer.tokenize(row['Description']), axis=1)
df['Description Tokenized'] = df['Description Tokenized'] + df['Keywords']
df.head(5)
```

```python
def remove_caps(s):
     s = [t.lower() for t in s]
     return s

df['Description Tokenized'] = df.apply(lambda row: remove_caps(row['Description Tokenized']), axis=1)
df.head(5)
```

```python
def remove_punctuation(tokens):
     tokens = [t for t in tokens if not t in list(string.punctuation)]
     return tokens

df['Description Tokenized'] = df.apply(lambda row: remove_punctuation(row['Description Tokenized']), axis=1)
df.head(5)
```

```python
def remove_stopwords(tokens):
     tokens = [t for t in tokens if not t in stopwords.words("english")]
     return tokens

df['Description Tokenized'] = df.apply(lambda row: remove_stopwords(row['Description Tokenized']), axis=1)
df.head(5)
```

```python
wnl = nltk.stem.WordNetLemmatizer()
def lemmatize(s):
     s = [wnl.lemmatize(word) for word in s]
     return s

df['Description Tokenized'] = df.apply(lambda row: lemmatize(row['Description Tokenized']), axis=1)
df.head(5)
```

```python
df = df.drop(columns=['Genre(s)','Keywords'])
```

```python
df.to_csv('../data/test.csv', index=False)
```
